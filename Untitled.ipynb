```python
import yfinance as yf
import pandas as pd
symbol = '^GSPC'
start_date = '2022-01-01'
end_date = '2025-01-01'
sp500_data = yf.download(symbol, start=start_date, end=end_date)
display(sp500_data.head())
```

```python
for col in ['Close', 'Open', 'Volume']:
    for i in range(1, 4):
        sp500_data[f'{col}_Lag{i}'] = sp500_data[col].shift(i)
```

```python
sp500_data['Close_RollingMean_5'] = sp500_data['Close'].rolling(window=5).mean()
sp500_data['Close_RollingMean_10'] = sp500_data['Close'].rolling(window=10).mean()
sp500_data['Close_RollingMean_20'] = sp500_data['Close'].rolling(window=20).mean()
display(sp500_data.head())
```

```python
sp500_data.dropna(inplace=True)
display(sp500_data.head())
```

```python
sp500_data['Close'].plot()
```

```python
train_sp500_data = sp500_data.iloc[:int(.99*len(sp500_data)), :]
test_sp500_data = sp500_data.iloc[int(.99*len(sp500_data)):, :]
```

```python
features = ['Open', 'Volume', 'Close_Lag1', 'Close_Lag2', 'Close_Lag3','Open_Lag1', 'Open_Lag2', 'Open_Lag3', 'Volume_Lag1', 'Volume_Lag2', 'Volume_Lag3',
'Close_RollingMean_5', 'Close_RollingMean_10', 'Close_RollingMean_20']
target = 'Close'
```

```python
x = sp500_data[features]
y = sp500_data[target]
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3,5,7],
    'subsample': [0.7, 0.8, 1,0],
    'colsample_bytree': [0.7, 0.8, 1.0]

}
```

```python
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
tscv = TimeSeriesSplit(n_splits=5)
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid,
                           cv=tscv, scoring='r2', verbose=2, n_jobs=-1)
print("Starting GridSearchCV... This may take a while.")
grid_search.fit(x, y)
print("GridSearchCV completed.")
```

```python
print("Best parameters found: ", grid_search.best_params_)
print("Best R-squared score: ", grid_search.best_score_)
```

```python
tscv = TimeSeriesSplit(n_splits=5)
accuracy_scores = []
x = sp500_data[features]
y = sp500_data[target]
for fold, (train_index, val_index) in enumerate(tscv.split(x)):
  print(f"\n-- Fold {fold+1} ---")
  x_train, x_val = x.iloc[train_index], x.iloc[val_index]
  y_train, y_val = y.iloc[train_index], y.iloc[val_index]
  model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100)
  model.fit(x_train,y_train)
  predictions = model.predict(x_val)
  accuracy = model.score(x_val, y_val)
  accuracy_scores.append(accuracy)
  print(f"R-squared for Fold {fold+1}: {accuracy:4f}")
  last_x_val = x_val
  last_y_val = y_val
  last_predictions = predictions
```

```python
print(f"\nAverage R-squared across all folds: {np.mean(accuracy_scores):.4f}")
print(f"Standard deviation of R-squared across all folds: {np.std(accuracy_scores):.4f}")
```

```python
model = xgb.XGBRegressor(objective='reg:squarederror ', **grid_search.best_params_, random_state=42)
model.fit(train_sp500_data[features], train_sp500_data[target])
```

```python
predictions = model.predict(test_sp500_data[features])
print('Model Predictions:')
print(predictions)
```
